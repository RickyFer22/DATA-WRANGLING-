**Esta función útil que permite eliminar columnas en un conjunto de datos que no tienen información en más del 50% de las filas. Este proceso de limpieza de datos ayuda a simplificar el conjunto de datos, mejorando su calidad y eficiencia en su posterior análisis. Con nuestra función, podemos automatizar este proceso y ahorrar tiempo en la limpieza de grandes conjuntos de datos."** 

```python
def eliminar_columnas_con_nulos(df):
    # Calculamos el porcentaje de valores nulos en cada columna del conjunto de datos
    null_porciento = 100*(df.isnull().sum()/len(df))
    
    # Seleccionamos solo las columnas que tienen más del 50% de valores nulos
    columnas_con_nulos = null_porciento[null_porciento > 50].index
    
    # Eliminamos las columnas seleccionadas del conjunto de datos
    df_sin_columnas_con_nulos = df.drop(columns=columnas_con_nulos)
    
    # Imprimimos las columnas eliminadas
    print("Columnas eliminadas:")
    print(columnas_con_nulos)
    
    return df_sin_columnas_con_nulos
```

```python
# Ingreso bank_ls_train en la función para eliminar las columnas 
bank_ls_train = eliminar_columnas_con_nulos(bank_ls_train)
```

**Esta funcion utiliza un bucle "for" que registra todas las columnas numéricas del dataframe y calcula la mediana de cada una de ellas. Luego, utilice el método "fillna" de pandas para reemplazar los valores faltantes con la mediana calculada previamente. El parámetro "inplace=True" hace que los cambios se realicen directamente en el dataframe original.**


**Por último, la función devuelve el marco de datos con los valores faltantes eliminados por la mediana.**
**Este código parece ser una solución simple y efectiva para el problema de valores faltantes en un marco de datos**

```python
# Funcion para reemplazar datos nulos por la MEDIANA
import pandas as pd

def fill_missing_with_median(df):
    for col in df.select_dtypes(include='number').columns:
        median = df[col].median()
        df[col].fillna(median, inplace=True)
    return df
```

```python
# Apliacamos para nuestro dataframe bank_ls_train
bank_ls_train = fill_missing_with_median(bank_ls_train)
```

### Outliers

**Como outliers en las variables numéricas, a veces no se deben eliminar porque pueden contener información importante, dependiendo de la variable numérica que estemos analizando. Por eso, a menudo construimos gráficas tipo "boxplot" de las columnas numéricas.**

**Visualizamos de otra manera las columnas float64**

```python

#La lista llamada "cols_num" que contiene los nombres de las columnas para las cuales se crearán los gráficos de caja.
cols_num = ["Current Loan Amount", "Credit Score","Annual Income",     
            "Monthly Debt", "Years of Credit History",
            "Number of Open Accounts",  
            "Number of Credit Problems","Current Credit Balance",
            "Maximum Open Credit", "Bankruptcies","Tax Liens"]


# El tamaño de la figura se establece en (8,30).
fig, axs = plt.subplots(nrows=len(cols_num), ncols=1, figsize=(8,30))

#La función "fig.subplots_adjust" para ajustar el espacio entre las filas del gráfico.
fig.subplots_adjust(hspace=0.5)

# El bucle "for" que itera sobre cada columna en "cols_num".
for i, col in enumerate(cols_num):
    sns.boxplot(data=bank_ls_train, x=col, ax=axs[i])
    axs[i].set_title(col)
```

**Para tratar los outliers voy utilizar el rango intercuartil (IQR) y multiplicarlo por un factor constante, por ejemplo, 1.5. El rango intercuartil es la diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1) de los datos. Los valores que están fuera de este rango se consideran outliers.**



```python
# Columnas a procesar
cols_num = ["Credit Score", "Monthly Debt", "Current Credit Balance", "Maximum Open Credit"]

# Definir los límites de los valores aceptables
limits = {}
for col in cols_num:
    Q1 = bank_ls_train[col].quantile(0.25)
    Q3 = bank_ls_train[col].quantile(0.75)
    IQR = Q3 - Q1
    limits[col] = (Q1 - 1.5 * IQR, Q3 + 1.5 * IQR)

# Filtrar los outliers
for col in cols_num:
    bank_ls_train = bank_ls_train.loc[(bank_ls_train[col] >= limits[col][0]) & (bank_ls_train[col] <= limits[col][1])]
```
